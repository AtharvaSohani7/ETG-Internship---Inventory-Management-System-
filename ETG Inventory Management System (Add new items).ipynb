```python
rec = {100: {'Product': 'Cadbury', 'Category': 'Chocolate', 'Price': 40, 'Quantity': 100, 'Expiry Date': '05/22'}}
```

# Adding items to the inventory


```python
id = str(input("Enter Product ID:"))
product = str(input("Enter Product:"))
cat = str(input("Enter Category:"))
pr = int(input("Enter Price:"))
qn = int(input("Enter Quantity:"))
exp = input("Enter expiry date:")

if id in rec:
    rec[id] = {'Product': product, 'Category': cat, "Price":pr ,'Quantity': (rec[id]["Quantity"]) + qn, "Expiry Date": exp }       
else:
    rec[id] = {'Product': product, 'Category': cat, "Price":pr ,'Quantity': qn, "Expiry Date": exp }

import json

js = json.dumps(rec)

fd = open("rec.json",'w')
fd.write(js)
fd.close()
```
