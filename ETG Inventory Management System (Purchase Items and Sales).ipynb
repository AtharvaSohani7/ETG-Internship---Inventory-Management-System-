# Purchasing items

```python
import time 
time.ctime()


id = int(input("Enter Product ID : ")) 
quant = int(input("Enter the quantity : "))
st = str(id)

for i in st:
    if st in rec:
        print("\nABC GENERAL STORES",)
        print("***********************")
        print("Product ID :",(st))
        print("Product :",(rec[st]["Product"]))
        print("Category: ",(rec[st]["Category"]))
        print("Price: ",(rec[st]["Price"]))
        print("Quantity: ",(quant))
        print("Expiry Date: ",(rec[st]["Expiry Date"])) 
        print("------------------------")
        print("Billing Amout: ", quant * (rec[st]["Price"]))
        (rec[st]["Quantity"]) = (rec[st]["Quantity"]) - quant
        print(time.ctime())
        print("***********************")
        break
    else: 
        print("Sorry, the product ID does not exist. Enter a valid product ID.")
        break
```
